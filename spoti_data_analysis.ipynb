```python
import pandas as pd
df1=pd.read_csv(r'clean_spoti_data.csv')
```

```python
#Top 5 artists
artist_popularity = df1.groupby('Artist_Name')['Followers'].max().sort_values(ascending=False)
artist_popularity=artist_popularity[:5]
df2=pd.DataFrame(artist_popularity)
df2.reset_index()
```

```python
#2. Genre analysis
genre_count_dict = {}

# Create a DataFrame
l = list(df1['Genre'].unique())
print(l)

c=pd.DataFrame(l)
c.columns=['Genre']
# Iterate through each row in the DataFrame and count genres
for row in c['Genre']:
    genres = eval(row)
    for genre in genres:
        if genre in genre_count_dict:
            genre_count_dict[genre] += 1
        else:
            genre_count_dict[genre] = 1

# Display the genre counts in the dictionary
for genre, count in genre_count_dict.items():
    print(f'{genre}: {count}')
```

```python
# 3. No of album releases per year

df1['Release date'] = pd.to_datetime(df1['Release date'])

# Extract the release year from the date
df1['Release Year'] = df1['Release date'].dt.year

# Group by release year and count the number of albums
album_counts = df1.groupby('Release Year').size().reset_index(name='Count')
album_counts
```
